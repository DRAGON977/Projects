```python
df_country=pd.read_excel('/content/drive/MyDrive/PROJECT/Zomatodataset/Country-Code.xlsx')
df_country.head()
```

```python
#finding missing value
df.columns
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head(3)
```

```python
final_df.columns
```

```python
final_df.Country.value_counts()
```

```python
country_names=final_df.Country.value_counts().index
```

```python
final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
#pie chart plotting
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
#plt.title("Observation:Zomato maximum records or transactions from India , USA and UK")
plt.annotate("Observation: Zomato maximum records or transactions from India, USA, and UK",
             xy=(0.3, -0.4),
             xycoords='axes fraction',
             ha='center',
             fontsize=10)

# Display the pie chart
plt.show()
```

```python
#relationships
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

```python
#finding relations wrt to different countries
#visulization
#conclusions
'''
1.when rating is between 4.5-4.9(excellent)
2.when rating is btwn 4.0-4.4(very good)
3.when rating is btwn 3.5-3.9(good)
4.when rating is btwn 2.5-3.4(average)
5.when rating is btwn 1.8-2.4(poor)
'''
```

```python
ratings
```

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(15,10)
sns.barplot(x="Aggregate rating", y="Rating Count", data=ratings)
```

```python
matplotlib.rcParams['figure.figsize']=(15,10)
sns.barplot(x="Aggregate rating", y="Rating Count", hue="Rating color", data=ratings,palette=['white','red','orange','yellow','green','green'])
```

```python
#observations
'''
1.Not rated count is very high
2.max number of ratings are between 2.5-3.4

'''
```

```python
#countplot:categorical plotting
sns.countplot(x="Rating color",data=ratings,palette=['white','red','orange','yellow','lightgreen','green'])
```
